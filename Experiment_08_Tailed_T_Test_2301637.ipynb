**Q1]What is 2 tailed and  tailed T Test ?  **

Ans:
A T-test is a statistical hypothesis test that is used to compare the means of two groups. The "tailed" aspect refers to the directionality of the hypothesis being tested:

a] One-tailed T-test:
 This test is used when you have a specific directional hypothesis. You are interested in whether the mean of one group is significantly greater than or significantly less than the mean of another group (or a hypothesized value). The critical region for rejecting the null hypothesis is entirely in one tail of the distribution (either the upper tail or the lower tail).

**b]Two-tailed T-test:**
 This test is used when you have a non-directional hypothesis. You are interested in whether there is any significant difference between the means of two groups (or a hypothesized value), regardless of the direction of that difference. The critical region for rejecting the null hypothesis is split between both tails of the distribution (half in the upper tail and half in the lower tail).



**Q2]Explain what is Type I and Type II errors with an example.**

Ans:
Type I and Type II errors are critical concepts in hypothesis testing:

**a] Type I Error (False Positive):** This occurs when you reject a true null hypothesis. In simpler terms, you conclude there is an effect or difference when, in reality, there isn't. The probability of making a Type I error is denoted by alpha (α), which is also the significance level of the test (commonly 0.05).

Example: Imagine a new drug is developed to cure a disease. The null hypothesis (H₀) is that the drug has no effect. A Type I error would be concluding that the drug is effective when, in fact, it has no real impact on the disease.

**b]Type II Error (False Negative): **
This occurs when you fail to reject a false null hypothesis. In simpler terms, you conclude there is no effect or difference when, in reality, there is one. The probability of making a Type II error is denoted by beta (β). The power of a test (1 - β) is the probability of correctly rejecting a false null hypothesis.

Example: Using the same drug example, a Type II error would be concluding that the drug has no effect when, in fact, it is effective at curing the disease.


**Q3]What is power of statistical test?**

Ans:


The power of a statistical test is the probability that the test will correctly reject a false null hypothesis. In simpler terms, it's the likelihood of finding an effect or difference when an effect or difference truly exists. It's often denoted as 1 - β (where β is the probability of a Type II error). A test with high power is more likely to detect a real effect.


**Q4]Explain input and out put characteristics for  stats.ttest_rel function in your own words**

Ans:
 Let's break down the stats.ttest_rel function from the perspective of its inputs and outputs:

Input Characteristics (What it expects):

The stats.ttest_rel function is specifically designed for paired samples. This means it needs two sets of numerical data where each observation in the first set is directly related to a corresponding observation in the second set. Think of it like 'before' and 'after' measurements on the same individuals, or measurements from two different treatments applied to matched pairs.

1.Two Arrays/Lists: You'll typically provide two equal-length arrays or lists of numbers. Let's say you have sample1 and sample2.

2.Paired Data: The crucial part is that sample1[i] and sample2[i] must represent a pair (e.g., the same person's score before and after an intervention).
   
3. Numerical Data: Both inputs must contain quantitative (numerical) data.

Output Characteristics (What it gives you):

The function returns two main values:

1.T-statistic: This is a single numerical value that tells you the size of the difference between the means of your two paired samples, relative to the variation within those differences. A larger absolute t-statistic suggests a bigger difference between the pairs.

2.P-value: This is a probability value. It tells you the likelihood of observing a difference as extreme as, or more extreme than, the one you found in your data, if there were truly no difference between the population means from which your samples were drawn (i.e., if the null hypothesis were true). A small p-value (commonly less than 0.05) is usually interpreted as evidence against the null hypothesis, suggesting that the observed difference is statistically significant.


**Q5]What is function in stats module for identifying power of a statistical test. Explain out put range of it and its significance.**

```python
from statsmodels.stats.power import TTestIndPower

# Parameters for power calculation
effect_size = 0.5  # Standardized effect size (Cohen's d)
alpha = 0.05       # Significance level
nobs1 = 20         # Number of observations in one sample

# Create a power analysis object
power_analyzer = TTestIndPower()

# Calculate the power
power = power_analyzer.solve_power(
    effect_size=effect_size,
    nobs1=nobs1,
    alpha=alpha,
    ratio=1.0,         # Ratio of sample sizes (nobs2/nobs1); 1.0 for equal sample sizes
    alternative='two-sided' # 'two-sided', 'smaller', or 'larger'
)

print(f"Calculated Power: {power:.4f}")
```

### Output Range and Significance

*   **Output Range:** The output of a power calculation function, like the `power` variable in the example above, will always be a value between **0 and 1 (inclusive)**. It is a probability.

*   **Significance:**
    *   **Close to 0:** A power close to 0 means your test has a very low chance of detecting a real effect, even if one exists. This implies a high risk of a Type II error (false negative).
    *   **Close to 1 (or typically 0.8 and above):** A power close to 1 (e.g., 0.8 or 80% is often considered a good target) means your test has a high probability of correctly detecting a real effect. This is desirable, as it means you are less likely to miss a true effect.

In essence, the power indicates the sensitivity of your statistical test to detect an effect of a certain size, given your sample size and chosen significance level. Higher power is generally preferred for more reliable research findings.

**Q3]Explain role of Critical value and p-value in Chi square test.**

Critical Value in Chi-square Test

1.What it is: The critical value is a predetermined threshold found in a Chi-square distribution table (or calculated by software) based on your chosen significance level (alpha, α) and the degrees of freedom of your test.

2. Its Role: It defines the rejection region(s) for the null hypothesis. If your calculated Chi-square test statistic falls into this rejection region (i.e., is greater than the critical value for a typical right-tailed Chi-square test), you reject the null hypothesis.

3.Think of it as: A fixed boundary. You compare your test statistic to this boundary. If it crosses the boundary, it's considered 'significant'.

P-value in Chi-square Test

1.What it is: The p-value is the probability of obtaining a test statistic as extreme as, or more extreme than, the one observed in your sample data, assuming the null hypothesis is true.

2.Its Role: It quantifies the strength of evidence against the null hypothesis. A small p-value suggests that your observed data would be very unlikely if the null hypothesis were true, thus providing evidence to reject the null hypothesis.

3.Think of it as: The probability of your results happening by chance if there was no real association or difference. The smaller the p-value, the less likely your results are due to chance.

How they work together (Decision Rule):

1. Using Critical Value: If the Chi-square Test Statistic > Critical Value, you reject the null hypothesis.
2. Using P-value: If the P-value < Significance Level (α), you reject the null hypothesis.

Both methods lead to the same conclusion, assuming a consistent significance level. The p-value is often preferred today because it provides more nuanced information about the strength of the evidence (e.g., a p-value of 0.001 is stronger evidence than 0.045, even if both lead to rejection at α=0.05), whereas the critical value only gives a pass/fail outcome.
