**Q1]What is need for Hypothesis Testing ?**

Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. It's crucial for several reasons:

1]Informed Decision Making: It helps in making objective decisions about a population parameter (e.g., mean, proportion) when only sample data is available.

2]Validating Claims or Assumptions: It allows researchers and analysts to formally test whether a claim or assumption about a population is supported by empirical evidence.

3]Evaluating Effectiveness: It's widely used to determine if a new treatment, intervention, marketing strategy, or product change has a statistically significant effect.

4]Distinguishing Real Effects from Chance: It helps to ascertain whether an observed difference or relationship in data is genuine (statistically significant) or merely due to random chance.
    
5]Scientific Research: It forms the backbone of the scientific method, enabling the testing of theories and the advancement of knowledge across various fields like medicine, economics, and social sciences.

### Important Steps in Hypothesis Testing

```mermaid
graph TD
    A[Start: Formulate Research Question] --> B{Formulate Null (H0) and Alternative (Ha) Hypotheses}
    B --> C[Choose a Significance Level (α)]
    C --> D[Select an Appropriate Statistical Test]
    D --> E[Collect Sample Data]
    E --> F[Calculate Test Statistic]
    F --> G[Determine P-value or Critical Value]
    G --> H{Compare P-value to α or Test Statistic to Critical Value}
    H -- P-value < α OR Test Stat. in Rejection Region --> I[Reject Null Hypothesis (H0)]
    H -- P-value ≥ α OR Test Stat. NOT in Rejection Region --> J[Fail to Reject Null Hypothesis (H0)]
    I --> K[Draw Conclusion: Support Ha]
    J --> K[Draw Conclusion: No Support for Ha]
    K --> L[End: Interpret Results in Context]
```

**Explanation of Steps:**

1.  **Formulate Research Question**: Clearly define what you want to investigate.
2.  **Formulate Null (H0) and Alternative (Ha) Hypotheses**:
    *   **Null Hypothesis (H0)**: A statement of no effect or no difference. It's what you assume to be true.
    *   **Alternative Hypothesis (Ha)**: A statement that contradicts the null hypothesis, representing what you want to prove.
3.  **Choose a Significance Level (α)**: Set the probability of rejecting the null hypothesis when it is actually true (Type I error). Common values are 0.05 or 0.01.
4.  **Select an Appropriate Statistical Test**: Based on the data type, distribution, and research question (e.g., t-test, ANOVA, chi-square).
5.  **Collect Sample Data**: Gather relevant data from the population.
6.  **Calculate Test Statistic**: Compute a value from the sample data that summarizes the evidence against the null hypothesis.
7.  **Determine P-value or Critical Value**:
    *   **P-value**: The probability of observing a test statistic as extreme as, or more extreme than, the one calculated, assuming the null hypothesis is true.
    *   **Critical Value**: The threshold value(s) that define the rejection region(s) for the null hypothesis.
8.  **Compare P-value to α or Test Statistic to Critical Value**:
    *   If P-value < α (or if the test statistic falls into the rejection region), you have enough evidence to reject H0.
    *   If P-value ≥ α (or if the test statistic does not fall into the rejection region), you do not have enough evidence to reject H0.
9.  **Make a Decision (Reject H0 or Fail to Reject H0)**:
10. **Draw a Conclusion**: State your findings in the context of your research question, explaining what the decision means.
11. **Interpret Results in Context**: Discuss the practical implications of your findings.

**Q3]What is a student t distribution?**

Ans:
 The Student's t-distribution, often referred to simply as the t-distribution, is a probability distribution that is used when estimating the mean of a normally distributed population in situations where the sample size is small and the population standard deviation is unknown. It's a fundamental concept in inferential statistics, particularly for hypothesis testing and constructing confidence intervals.

Here are its key characteristics:

1]Shape: It is bell-shaped and symmetric, similar to the normal distribution, but with heavier tails. This means it has more probability in its tails compared to a normal distribution, making it more prone to producing values that fall far from the mean.

2] Degrees of Freedom (df): The shape of the t-distribution is determined by a single parameter called 'degrees of freedom'. As the degrees of freedom increase, the t-distribution approaches the normal distribution. Generally, for degrees of freedom greater than about 30, the t-distribution is very similar to the standard normal (Z) distribution.
  
3]Application: It is commonly used for:

a)Hypothesis testing: To test hypotheses about a population mean when the sample size is small and the population standard deviation is unknown (e.g., one-sample t-test, independent samples t-test, paired samples t-test).
b)Confidence intervals: To construct confidence intervals for a population mean under the same conditions.
4]Origin: It was developed by William Sealy Gosset in 1908, who published under the pseudonym 'Student' while working at Guinness Brewery, as he was not allowed to publish research under his own name.

Q4]What is level of significance ? What role t plays in hypothesis Testing



Ans:
 The level of significance (α), also known as the alpha level, is a crucial concept in hypothesis testing. It represents the probability of rejecting the null hypothesis when it is actually true (Type I error).

Here's a breakdown:

    Definition: It's a threshold set by the researcher before conducting the test. If the p-value of the test statistic is less than the significance level, the null hypothesis is rejected.
    Common Values: Commonly used significance levels are 0.05 (5%), 0.01 (1%), and 0.10 (10%).
    Interpretation: A significance level of 0.05 means there is a 5% risk of concluding that a difference or relationship exists when there is no actual difference or relationship in the population.
    Purpose: It helps researchers control the probability of making a Type I error and provides a clear criterion for making a decision about the null hypothesis.

The 't' in hypothesis testing typically refers to the t-statistic or the Student's t-distribution.

Its role in hypothesis testing is significant, especially when dealing with small sample sizes or when the population standard deviation is unknown:

1]T-statistic: This is a calculated value from sample data that measures how many standard errors the sample mean is away from the hypothesized population mean. It's essentially a signal-to-noise ratio: the larger the absolute value of the t-statistic, the stronger the evidence against the null hypothesis.

2]Student's t-distribution: Once the t-statistic is calculated, it is compared against critical values from the t-distribution (or used to calculate a p-value from the t-distribution) to make a decision about the null hypothesis. The shape of the t-distribution depends on the degrees of freedom, becoming more like a normal distribution as the degrees of freedom increase.

3]Application in Tests: The t-distribution is fundamental to various t-tests:
  
a]One-sample t-test: To test if a sample mean is significantly different from a known or hypothesized population mean.

b]Independent samples t-test: To compare the means of two independent groups.
      
c]Paired samples t-test: To compare means from the same group at different times or under different conditions.

In essence, the t-statistic quantifies the evidence against the null hypothesis, and the t-distribution provides the framework for evaluating the probability of observing such evidence by chance, thereby guiding the decision to reject or fail to reject the null hypothesis.

***

### Formula for One-Sample t-statistic

The t-statistic for a one-sample test is calculated using the following formula:

$$ t = \frac{\bar{x} - \mu_0}{\frac{s}{\sqrt{n}}} $$

Where:

*   **$\bar{x}$ (sample mean)**: This is the average age calculated from the sample of students you collect from the TE IT class. It represents the central tendency of your collected data.
*   **$\mu_0$ (hypothesized population mean)**: This is the value from your null hypothesis. In our case, it's 19 years, as we are testing if the average age of students in the TE IT class is 19.
*   **$s$ (sample standard deviation)**: This is a measure of the spread or variability of the ages within your sample of students. It indicates how much the individual ages in your sample deviate from the sample mean.
*   **$n$ (sample size)**: This is the total number of students included in your sample from the TE IT class. A larger sample size generally leads to a more reliable estimate of the population mean.

### Degrees of Freedom (df)

For a one-sample t-test, the degrees of freedom (df) are calculated as:

$$ df = n - 1 $$

Where 'n' is the sample size. The degrees of freedom are important because they determine the specific shape of the t-distribution, which is used to find the p-value or critical values for making a decision about the null hypothesis.

#### Instructions

1.  **Determine the P-value:**

    *   **Using T-Distribution Table/Software:** Once you have the calculated t-statistic and the degrees of freedom (df = n - 1, where 'n' is the sample size), you can look up the p-value in a t-distribution table or use statistical software (like Python's `scipy.stats.t.sf` for survival function or `scipy.stats.t.cdf` for cumulative distribution function). The p-value represents the probability of observing a test statistic as extreme as, or more extreme than, the one calculated, assuming the null hypothesis is true.
    *   **For a Two-Tailed Test (Hₒ: μ ≠ 19):** If your alternative hypothesis is two-tailed (e.g., the average age is *not equal to* 19), the p-value is the sum of the probabilities in both tails of the t-distribution. You find the probability of getting a t-statistic greater than the absolute value of your calculated t-statistic AND the probability of getting a t-statistic less than the negative absolute value. Often, statistical software directly provides this two-tailed p-value.

2.  **Determine Critical T-values:**

    *   **Using T-Distribution Table:** Alternatively, you can find the critical t-values for your chosen significance level (α = 0.05) and degrees of freedom (df = n - 1). For a two-tailed test, the significance level α is split into two tails (α/2 on each side).
    *   **For a Two-Tailed Test (α = 0.05):** You would look for the t-value that leaves 0.025 (α/2) probability in the upper tail, and another t-value that leaves 0.025 probability in the lower tail. This will give you two critical values: a positive one (e.g., tₐ/₂) and a negative one (e.g., -tₐ/₂). These critical values define the 'rejection regions' of the distribution. If your calculated t-statistic falls outside these critical values (i.e., in either rejection region), you reject the null hypothesis.
